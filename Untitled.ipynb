```python
def buy(curPrice,curMoney=200, tax=0):
    curChunk = (curMoney-(tax*curMoney))/curPrice
    return curChunk
```

```python
def sell(curPrice, curMoney=200, tax=0):
    curChunk = (curMoney-(tax*curMoney))/curPrice
    return curChunk
```

```python
df = pd.read_csv('050026.csv')
    # dataValue = data.values
data = df.values
value = data[:,2]

times = 0
cost = 0
cash = 100000
chunk = 0
exchange = 0
for cur in value:
    # 第一次交易
    if times == 0:
        curChunk = buy(cur)
        cost += 200
        cash -= 200
        chunk += curChunk
        times += 1
        exchange += 1

    # 计算收益情况
    income = chunk*cur - cost
    # 计算收益比例
    incomePercent = 0 if(cost == 0) else income / cost
    if incomePercent > 0.1:
        # sell
        cost -= 200
        money += 200
        curChunk = sell(cur)
        chunk -= curChunk
        exchange += 1
    elif incomePercent < -0.05 or incomePercent == 0:
        # buy
        curChunk = buy(cur)
        cost += 200
        cash -= 200
        chunk += curChunk 
        exchange += 1
    print('交易次数:'+str(exchange)+'收益是:'+str(income))
    
cur = value[len(value)-1]
total = cash + chunk * cur
print(total)
```
