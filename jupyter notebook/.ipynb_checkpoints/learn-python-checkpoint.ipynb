## 解一元二次方程

```python
import math
```

```python
def quadratic(a, b, c):
    if a != 0:
        return (-b+math.sqrt((b**2)-(4*a*c)))/(2*a),(-b-math.sqrt((b**2)-(4*a*c)))/(2*a)
    else:
        return -c/b
```

```python
answer = quadratic(1, 3, -4)
```

```python
print(answer)
```

## 可变参数与关键字参数
*args  *kw

```python
def mul(*args):
    res = 1
    for num in args:
        res = res * num
    return res
```

```python
mul(5)
```

```python
mul(5,6)
```

```python
mul(5,6,7)
```

```python
mul(5,6,7,9)
```

```python
def fact(n):
    if n == 1:
        return 1
    return n * fact(n-1)
```

```python
fact(10)
```

## 去除字符串前后可能存在的空格

```python
def trim(string):
    if(string == ''):
        return ''
    if(string[0] == ' '):
        return trim(string[1:])
    elif(string[-1] == ' '):
        return trim(string[:-2])
    else:
        return string
```

```python
if trim('hello  ') != 'hello':
    print('测试失败!')
elif trim('  hello') != 'hello':
    print('测试失败!')
elif trim('  hello  ') != 'hello':
    print('测试失败!')
elif trim('  hello  world  ') != 'hello  world':
    print('测试失败!')
elif trim('') != '':
    print('测试失败!')
elif trim('    ') != '':
    print('测试失败!')
else:
    print('测试成功!')
```

# generator 生成器 
.next()
for 遍历

```python
def triangles(max):
    row = 0
    
    while(row < max):
        res = []
        for i in range(row):
            res.append(1)
        yield res
        row += 1
```

```python
g = triangles(1)
for i in g:
    print(i)
```
